```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd

# Function to scrape data
def scrape_info(url):
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Request failed for {url}: {e}")
        return {}

    soup = BeautifulSoup(response.text, 'html.parser')

    # Define sections with correct mapping to your dataset column names
    section_map = {
        "Overview": "Overview",
        "Symptoms": "Symptoms",
        "When to see a doctor": "When to See Doctor",
        "Causes": "Causes",
        "Risk factors": "Risk Factor",
        "Complications": "Complications",
        "Prevention": "Preventions"
    }

    content = {}
    for section, col_name in section_map.items():
        section_div = soup.find('section', {'aria-labelledby': section.lower().replace(" ", "-")})
        if section_div:
            text_content = []
            for element in section_div.find_all(['p', 'ul', 'li']):
                text_content.append(element.get_text(strip=True, separator=" "))
            unique_text = list(dict.fromkeys(text_content))
            content[col_name] = " ".join(unique_text) if unique_text else "No data found"
        else:
            content[col_name] = "No data found"

    return content

# Columns to check for missing values
columns_to_fill = ["Overview", "Symptoms", "When to See Doctor", "Causes", 
                   "Risk Factor", "Complications", "Preventions"]

# Iterate over rows and fill missing values
for index, row in tqdm(df_sample.iterrows(), total=len(df_sample), desc="Filling Missing Values"):
    if pd.notna(row["Link"]):  # Ensure there's a valid link
        scraped_data = scrape_info(row["Link"])

        for col in columns_to_fill:
            if pd.isna(row[col]) and col in scraped_data:
                df_sample.at[index, col] = scraped_data[col]

print("Missing values filled successfully!")
```
